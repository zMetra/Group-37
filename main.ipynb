```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
neighbors = gpd.read_file('Neighborhoods_Files/Neighborhoods_.shp')
trails = gpd.read_file('Trails/Parks_PARKS_OWNER_Trails.shp')
fields = gpd.read_file('Ballfields/Ballfields_PGH.shp')
parks = gpd.read_file('Ballfields/Ballfields_PGH.shp')
```

```python
FG        = '#ffffff'
BG        = '#000000'
BLACK     = '#000000'
DARK_BLUE = '#141B4D'
DARK_GOLD = '#B58500'
PITT_GOLD = '#FFB81C'
GREEN     = '#22ff22'
PAD       = .5
WPAD      = 1
HPAD      = 1
```

```python
fig, ax = plt.subplots(figsize = (10,10))
ax.set_facecolor(DARK_BLUE)
neighbors.plot(ax=ax, color=PITT_GOLD, edgecolor=DARK_GOLD)
fields.plot(ax=ax, edgecolor=BLACK, marker='v', color=GREEN)
fig.patch.set_facecolor(BG)
ax.yaxis.set_tick_params(color=FG, labelcolor=FG)
ax.xaxis.set_tick_params(color=FG, labelcolor=FG)
plt.title("Baseball fields around the City of Pittsburgh", color=FG)
plt.tight_layout(pad=PAD, w_pad=WPAD, h_pad=HPAD)
for spine in ax.spines.values():
    spine.set_color(FG)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10,10))
ax.set_facecolor("#141B4D")
neighbors.plot(ax=ax, color='#FFB81C', edgecolor='#B58500')
parks.plot(ax=ax, edgecolor='black', color='#22ff22')
plt.show()
```
